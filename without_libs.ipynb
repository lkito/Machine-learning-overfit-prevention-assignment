```python
train_x = train.drop(['id', 'target'], axis=1)
train_y = train['target']
test_x = test.drop(['id'], axis=1)
train_x
```

```python
# Scale features by normalization to avoid bias towards columns
def scaleFeature(dataFrame, column, scale):
    stdVal = np.nanstd(scale[column])
    meanVal = np.mean(scale[column])
    dataFrame[column] = (dataFrame[column] - meanVal)/stdVal
```

```python
for col in test_x.columns:
    scaleFeature(test_x, col, train_x)
for col in train_x.columns:
    scaleFeature(train_x, col, train_x)
train_x
```

```python
train_x.insert( 0, 'Ones', 1)
test_x.insert( 0, 'Ones', 1)
train_x
```

```python
# THESE FUNCTIONS WERE (mostly) TAKEN FROM SEMINAR CODE EXAMPLES AND THEN MODIFIED!!!
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def costReg(theta, X, y, learningRate):
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    first = np.multiply(-y, np.log(sigmoid(X * theta.T)).T)
    second = np.multiply((1 - y), np.log(1 - sigmoid(X * theta.T)).T)
    reg = (learningRate / (2 * len(X))) * np.sum(np.power(theta[:,1:theta.shape[1]], 2))
    return np.sum(first - second) / (len(X)) + reg
```

```python
def gradientReg(theta, X, y, learningRate):
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    
    parameters = int(theta.ravel().shape[1])
    grad = np.zeros(parameters)

    error = sigmoid(X * theta.T).T - y

    for i in range(parameters):

        term = np.multiply(error, X[:,i].T)
        
        if (i == 0):
            grad[i] = np.sum(term) / len(X)
        else:
            grad[i] = (np.sum(term) / len(X)) + ((learningRate / len(X)) * theta[:,i])
    
    return grad
```

```python
theta = np.zeros(301)
X = np.array(train_x.values)
y = np.array(train_y.values)
learningRate = 5
```

```python
costReg(theta, X, y, learningRate)
```

```python
gradientReg(theta, X, y, learningRate)
```

```python
from scipy.optimize import check_grad
print(check_grad(costReg, gradientReg, theta, X, y, learningRate))
```

```python
result = opt.fmin_tnc(func=costReg, x0=theta, fprime=gradientReg, args=(X, y, learningRate))
result
```

```python
costReg(result[0], X, y, learningRate)
```

```python
result[0]
```

```python
def predict(theta, X):
    probability = sigmoid(X.dot(theta.T))
    return probability
```

```python
theta_min = np.matrix(result[0])
print(theta_min.shape, test_x.shape)
predictions = predict(theta_min, test_x)
predictions
```

```python
# Get the values ready to write into file
test_y = pd.DataFrame(predictions)
indArr = []
for i in range(250, 20000):
    indArr.append(i)
test_y.insert(0, 'id', indArr, 1)
test_y.columns = ['id', 'target']
```
